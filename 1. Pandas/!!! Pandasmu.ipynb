# 4. Manipulasi Data

### 4.1 Bikin kolom baru berupa jumlah dari kolom lain

```python
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
```

```python
df.head(2)
```

```python
#cara lebih cepat
#axis = 1 == njumlah secara horizontal
df['Total'] = df.iloc[:,4:10].sum(axis=1)

df.head(2)
```

### 4.2 Hapus Kolom

```python
df = df.drop(columns=['Total'])
df.head(2)
```

### 4.3 Reordering Colum

```python
cols = list(df.columns.values)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
```

### 6.1. Memfilter string tertentu

```python
# mencari yang ada 'Mega' nya
df.loc[df['Name'].str.contains('Mega')].head(3)
```

```python
# mencari yang GAK ada 'Mega' nya (~)
df.loc[~df['Name'].str.contains('Mega')].head(3)
```

### 6.2. Library Regular Expression (re)

```python
import re

#yang mengandung Fire atau Grass (case sensitive)
df.loc[df['Type 1'].str.contains('Fire|Grass', regex=True)]
```

```python
#yang mengandung Fire atau Grass tapi ignore case
df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)]
```

```python
#nyari yang  awalannya pi... diikuti dengan potongan kata
# ^ awal kata
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)]
```

### 6.3 Perubahan Kondisional (mengganti nilai)

## 7. Aggregate Statistics (Groupby) : Data Analysis sederhana

### 7.1. Mean

```python
#melihat mean berdasarkan Type 1 dan sort by Attack (atau var lain karek diganti2 ae) dari yang tertinggi
df.groupby(['Type 1']).mean().sort_values('Attack', ascending=False)
```

### 7.2. Sum

```python
df.groupby(['Type 1']).sum()
```

### 7.3. Count

```python
df.groupby(['Type 1']).count()
```

```python
# melihat count dari kolom indeks 1 (Type 1) 
df['count'] = 1
df.groupby(['Type 1']).count()['count']
```

```python
# melihat count dari kolom indeks 1 dan 2 (Type 1 dan Type 2) 
df['count'] = 1
df.groupby(['Type 1', 'Type 2']).count()['count']
```

## 8. Working with large Amounts of Data

```python
df = pd.read_csv('Modified.csv')
```

### 8.1. Ngeload chunk per chunk

```python
new_df = pd.DataFrame(columns = df.columns)

'''
# 5  c h o n k  at a time (5 rows masuk satu row)
for df in pd.read_csv('Modified.csv', chunksize=5):
    print("CHUNK DF")
    print(df)
'''

for df in pd.read_csv('Modified.csv', chunksize=5):
    results = df.groupby(['Type 1']).count()
    
    new_df = pd.concat([new_df, results])
    
new_df
```
